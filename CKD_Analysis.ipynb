1. Patient Statistics
Write a program to calculate and display:

•	Total number of patients.

•	Total number of CKD patients.

•	Total number of Non-CKD patients.

•	Percentage of CKD patients rounded to two decimal places.

10. Kidney Health Risk Score

Assign one risk point for each condition:

•	Blood Pressure > 140

•	Blood Sugar > 180

•	Serum Creatinine > 2.0

•	eGFR < 60

•	Albumin ≥ 3

Calculate every patient's total risk score (0–5) and display the patient(s) having the highest risk score.


```python
highest_score = 0

risk_scores = []

for patient in patients:

    score = 0

    if patient[1] > 140:
        score += 1

    if patient[2] > 180:
        score += 1

    if patient[3] > 2.0:
        score += 1

    if patient[5] < 60:
        score += 1

    if patient[6] >= 3:
        score += 1

    risk_scores.append(score)

    if score > highest_score:
        highest_score = score

print("Highest Risk Score :", highest_score)

print("Patients with Highest Risk Score")

for i in range(len(patients)):

    if risk_scores[i] == highest_score:

        print(patients[i], "Risk Score =", risk_scores[i])
```

14. Data Validation

Check the entire dataset and identify invalid records based on the following rules:

•	No value should be negative.

•	Albumin must be between 0 and 5.

•	CKD must contain only 0 or 1.

Display all invalid records. If none exist, print an appropriate message.


```python
invalid = []

for patient in patients:

    valid = True

    for value in patient:

        if value < 0:
            valid = False

    if patient[6] < 0 or patient[6] > 5:
        valid = False

    if patient[7] != 0 and patient[7] != 1:
        valid = False

    if valid == False:
        invalid.append(patient)

if len(invalid) == 0:

    print("No Invalid Records Found.")

else:

    print("Invalid Records")

    for patient in invalid:
        print(patient)
```

```python
total_bp = 0
total_bs = 0
total_sc = 0
total_hb = 0
total_egfr = 0
total_age = 0

ckd = 0

egfr60 = 0

highest_score = 0

high_risk = 0

for patient in patients:

    total_age += patient[0]
    total_bp += patient[1]
    total_bs += patient[2]
    total_sc += patient[3]
    total_hb += patient[4]
    total_egfr += patient[5]

    if patient[7] == 1:
        ckd += 1

    if patient[5] < 60:
        egfr60 += 1

    if patient[1] > 150 and patient[2] > 180 and patient[3] > 2.5 and patient[5] < 45:
        high_risk += 1

    score = 0

    if patient[1] > 140:
        score += 1

    if patient[2] > 180:
        score += 1

    if patient[3] > 2.0:
        score += 1

    if patient[5] < 60:
        score += 1

    if patient[6] >= 3:
        score += 1

    if score > highest_score:
        highest_score = score

print("\nKidney Health Summary Report\n")

print("Total Patients :", len(patients))

print("CKD Patients :", ckd)

print("Non CKD Patients :", len(patients)-ckd)

print("Average Age :", round(total_age/len(patients),2))

print("Average Blood Pressure :", round(total_bp/len(patients),2))

print("Average Blood Sugar :", round(total_bs/len(patients),2))

print("Average Serum Creatinine :", round(total_sc/len(patients),2))

print("Average Hemoglobin :", round(total_hb/len(patients),2))

print("Average eGFR :", round(total_egfr/len(patients),2))

print("Percentage of CKD Patients :", round((ckd/len(patients))*100,2),"%")

print("Number of High Risk Patients :", high_risk)

print("Patients with eGFR <60 :", egfr60)

print("Highest Risk Score :", highest_score)

print("\nPatients having Highest Risk Score\n")

for patient in patients:

    score = 0

    if patient[1] > 140:
        score += 1

    if patient[2] > 180:
        score += 1

    if patient[3] > 2.0:
        score += 1

    if patient[5] < 60:
        score += 1

    if patient[6] >= 3:
        score += 1

    if score == highest_score:
        print(patient)
```
